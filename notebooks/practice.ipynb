```python
import matplotlib.pyplot as plt
import numpy as np
```

# Generación de datos
### Genera dos distribuciones aleatorias de un tamaño de 1.000 observaciones cada una:

#### 1. Una **Distribución Normal (Gaussiana)** con media y desvío estándar a elección.

```python
media_n = 100
desviacion = 15
muestra = 1000

distribucion_n = np.random.normal(media_n, desviacion, muestra)
distribucion_n
```

#### 2. Una **Distribución Uniforme** dentro de un rango que definas.

```python
limite_inferior = 25
limite_superior = 100
muestra = 1000

distribucion_u = np.random.uniform(limite_inferior, limite_superior, muestra)
distribucion_u
```

# Análisis estadístico

### Calcula para cada distribución los siguientes valores:
- Media (mean)
- Mediana (median)
- Varianza (var)
- Desvío estándar (std)
- Mínimo y máximo (min, max)

### Cálculos para Distribución Normal...

```python
print('Distribución Normal: ')
media = np.mean(distribucion_n)
mediana = np.median(distribucion_n)
varianza = np.var(distribucion_n)
desvio_est = np.std(distribucion_n)
minimo = np.min(distribucion_n)
maximo = np.max(distribucion_n)

print(f'Media: {media}')
print(f'Mediana: {mediana}')
print(f'Varianza: {varianza}')
print(f'Desvío Estandar: {desvio_est}')
print(f'Minimo: {minimo}')
print(f'Maximo: {maximo}')
```

### Cálculos para Distribución Uniforme...


```python
print('Distribución Uniforme: ')
media_u = np.mean(distribucion_u)
mediana_u = np.median(distribucion_u)
varianza_u = np.var(distribucion_u)
desvio_est_u = np.std(distribucion_u)
minimo_u = np.min(distribucion_u)
maximo_u = np.max(distribucion_u)

print(f'Media: {media_u:.2f}')
print(f'Mediana: {mediana_u:.2f}')
print(f'Varianza: {varianza_u:.2f}')
print(f'Desvío Estandar: {desvio_est_u:.2f}')
print(f'Minimo: {minimo_u:.2f}')
print(f'Maximo: {maximo_u:.2f}')
```

# Visualización

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(13, 8))

# Histograma de distribución normal
ax1.hist(distribucion_n, bins='sqrt', color='blue', edgecolor='white')
ax1.set_title('Distribución Normal')
ax1.set_xlabel('Valor')
ax1.set_ylabel('Frecuencia')
ax1.grid()

# Histograma de distribución uniforme
ax2.hist(distribucion_u, bins='sqrt', color='green', edgecolor='white')
ax2.set_title('Distribución Uniforme')
ax2.set_xlabel('Valor')
ax2.set_ylabel('Frecuencia')
ax2.grid()

fig.suptitle('Comparación de Distribuciones')
plt.tight_layout()
plt.show()
```

### Reflexión sobre las Distribuciones...

La **Distribución Normal** muestra una forma de campana, centrada en la media.

En cambio, la **Distribución Uniforme** presenta una frecuencia constante. No se observa concentración en ningún valor específico.

La distribución normal representa fenómenos donde los valores tienden a agruparse cerca de un promedio por causas naturales o acumulativas. La uniforme, en cambio, muestra situaciones donde todos los valores dentro de un rango son igualmente probables, como por ejemplo en procesos aleatorios.
